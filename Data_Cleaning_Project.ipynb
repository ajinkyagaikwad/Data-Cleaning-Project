**Getting rid of leading spaces in 'Club' column**

```python
fifa['Club'] = fifa['Club'].str.strip()
```

**Formatting the 'Contract' column by creating 3 new columns as - Contract start, contract end, and contract length(years)**

```python
def extract_contract_info(contract):
	if contract == 'Free' or 'On Loan' in contract:
		start_date = np.nan
		end_date = np.nan
		contract_length = 0
	else:
		start_date, end_date = contract.split(' ~ ')
		start_year = int(start_date[:4])
		end_year = int(end_date[:4])
		contract_length = end_year - start_year
	return start_date, end_date, contract_length
```

**Applying the function created for formatting of 'Contract' column to the dataframe**

```python
new_cols = ['Contract start', 'contract end', 'contract length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))
```

**Inserting new columns to the dataframe 'Fifa'**

```python
for i in range(len(new_cols)):
	fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value=new_data[i])
```

```python
fifa.sample(3)
```

**Insert a new column 'Contract Status' based on the status of the contract fetched from 'Contract' column**

```python
def categorize_contract_status(contract):
	if contract == 'Free':
		return 'Free'
	elif 'On Loan' in contract:
		return 'On Loan'
	else:
		return 'Contract'
```

```python
fifa.insert(fifa.columns.get_loc('Contract')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
```

```python
fifa.sample(3)
```

**Converting the 'Height' column into uniform values in centimeters**

```python
def convert_height(height):
	if 'cm' in height:
		return int(height.strip('cm'))
	else:
		feet, inches = height.split("'")
		total_inches = int(feet)*12 + int(inches.strip('"'))
		return round(total_inches*2.54)
```

```python
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

**Renaming the 'Height' column as 'Height(cms)'**

```python
fifa = fifa.rename(columns={'Height':'Height(cms)'})
```

```python
fifa.sample(3)
```

```python
fifa['Weight'].unique()
```

**Converting weights into a uniform unit of 'kgs' and renaming the 'Weight' column as 'Weight(kgs)'**

```python
def convert_weight(weight):
	if 'kg' in weight:
		return int(weight.strip('kg'))
	else:
		pounds = int(weight.strip('lbs'))
		return round(pounds/2.205)
```

```python
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
```

```python
fifa = fifa.rename(columns={'Weight':'Weight(kgs)'})
```
